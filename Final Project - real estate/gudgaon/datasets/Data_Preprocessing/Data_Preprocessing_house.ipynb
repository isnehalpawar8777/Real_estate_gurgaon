# Data Cleaning

#### society 

```python
df.society.value_counts()
```

- these are houses so most of them are independent
- will replace nan with independent

```python
df.society.value_counts().shape
# we will clean the data using regular expression and then observe shape 
```

```python
# we are removing ratings from str
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
# 602 unique sociteties are there in out dataset
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

```python
df.head()
```

#### price

```python
df.price.value_counts()
```

- will change the name of column to price_cr
- there is one value 'Price on Request' will remove it
- will make all values in one unit cr
- will convert the dtype into float
- will check null values and deal with it 

```python
df.rename(columns = {'price':'price_cr'},inplace =True)
```

```python
df.head()
```

```python
df = df[df.price_cr != 'Price on Request']
# removed all the price on request rows
```

```python
df.shape
```

```python
# will make all values in one unit cr
def price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)

# this funtion will take string as input and return value in cr if it is Lac
# and also it will convert the dtype to float 
```

```python
df['price_cr'] = df.price_cr.str.split(' ').apply(price)
```

```python
df.head()
```

```python
# will check null values and deal with it 
df.price_cr.isnull().sum()
```

```python
# lets check null values 
df[df.price_cr.isnull()]
# it seems that there are too many null values 
# will handle this later 
```
